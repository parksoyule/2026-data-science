# **확률밀도함수 + 정규분포**

```python
sns.histplot(df['2024'], kde = True, stat='density')

mu, std = stats.norm.fit(df['2024'])
print(mu)
print(std)

x = np.linspace(df['2024'].min(), df['2024'].max(), 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x,p,'red')
```
