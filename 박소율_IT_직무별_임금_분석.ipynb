```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. 코랩용 한글 폰트 적용
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False

# 2. 데이터 세트 생성 (이미지 데이터 기준)
data = {
    '직무': ['IT기획자', 'IT컨설턴트', '업무분석가', '데이터분석가', 'IT PM',
           'IT아키텍트', 'UI/UX기획/개발자', 'UI/UX디자이너', '응용SW개발자',
           '시스템SW개발자', '정보시스템운용자', 'IT지원기술자', 'IT마케터',
           'IT품질관리자', 'IT테스터', 'IT감리', '정보보안전문가'],
    '2022': [419656, 476007, 544972, 347476, 463265, 518084, 291414, 217843, 311962, 247590, 326653, 190219, 378726, 402626, 208959, 456540, 362961],
    '2023': [481654, 476404, 532956, 380190, 456225, 534135, 315898, 224150, 341404, 278819, 435608, 215203, 483647, 442826, 189146, 485624, 454136],
    '2024': [562993, 471166, 436765, 376271, 443955, 492609, 326566, 251272, 337061, 296070, 492943, 245535, 536729, 470490, 173328, 502494, 478500]
}

df = pd.DataFrame(data)

# ==========================================
# 활동 1. 연도별 대푯값 추이 및 2024년 집중 분석
# ==========================================
print("📊 [연도별 IT 직무 임금 대푯값 변화]")
print("-" * 50)

# 1) 2022~2024년 평균과 중앙값 변화 출력하기
for year in ['2022', '2023', '2024']:
    mean_val = df[year].mean()
    median_val = df[year].median()
    print(f"✅ {year}년 - 평균: {mean_val:,.0f}원 / 중앙값: {median_val:,.0f}원")

print("-" * 50)

# 2) 2024년 기준 세부 분석 (최빈값 및 최대/최소)
mean_2024 = df['2024'].mean()
median_2024 = df['2024'].median()
mode_2024 = df['2024'].mode()

print(f"✅ 2024년 최빈값(Mode): \n{mode_2024.to_string(index=False)} \n(※ 수치형 데이터라 중복값이 없어 모두 출력됨)")
print("-" * 50)

# 가장 높은 임금과 가장 낮은 임금 찾기 (2024년)
max_job = df.loc[df['2024'].idxmax(), '직무']
max_wage = df['2024'].max()
min_job = df.loc[df['2024'].idxmin(), '직무']
min_wage = df['2024'].min()

print(f"💰 2024년 최고 임금 직무: {max_job} ({max_wage:,.0f}원)")
print(f"📉 2024년 최저 임금 직무: {min_job} ({min_wage:,.0f}원)")
print(f"격차: {(max_wage - min_wage):,.0f}원\n")


# ==========================================
# 활동 2. 데이터 시각화 (그래프 그리기)
# ==========================================

# 세 개의 그래프를 나란히 보여주기 위해 가로 길이를 18로 늘립니다.
plt.figure(figsize=(18, 6))

# [그래프 1] 막대그래프 (2024년 직무별 임금 비교)
plt.subplot(1, 3, 1)
df_sorted = df.sort_values(by='2024', ascending=True) # 보기 쉽게 오름차순 정렬
plt.barh(df_sorted['직무'], df_sorted['2024'], color='skyblue', edgecolor='black')
plt.title('2024년 IT 직무별 임금 (막대그래프)', fontsize=14)
plt.xlabel('임금', fontsize=12)
plt.ylabel('직무', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# [그래프 2] 히스토그램 (2024년 임금 데이터 분포 확인)
plt.subplot(1, 3, 2)
plt.hist(df['2024'], bins=6, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(mean_2024, color='red', linestyle='dashed', linewidth=2, label=f'24년 평균 ({mean_2024:,.0f})')
plt.axvline(median_2024, color='blue', linestyle='solid', linewidth=2, label=f'24년 중앙값 ({median_2024:,.0f})')
plt.title('2024년 IT 직무 임금 분포 (히스토그램)', fontsize=14)
plt.xlabel('임금 구간', fontsize=12)
plt.ylabel('직무 개수(빈도수)', fontsize=12)
plt.legend()

# [그래프 3] 선 그래프 (2022~2024년 대푯값 추이 확인)
# 22, 23, 24년도의 대푯값이 어떻게 변해왔는지 보여줍니다.
plt.subplot(1, 3, 3)
years = ['2022', '2023', '2024']
means = [df[y].mean() for y in years]
medians = [df[y].median() for y in years]

plt.plot(years, means, marker='o', color='red', linestyle='-', linewidth=2, label='평균(Mean)')
plt.plot(years, medians, marker='s', color='blue', linestyle='--', linewidth=2, label='중앙값(Median)')

# 그래프 위에 값(텍스트) 표시
for i, txt in enumerate(means):
    plt.text(years[i], means[i] + 5000, f'{txt:,.0f}', ha='center', va='bottom', fontsize=10, color='red')
for i, txt in enumerate(medians):
    plt.text(years[i], medians[i] - 10000, f'{txt:,.0f}', ha='center', va='top', fontsize=10, color='blue')

plt.title('연도별 임금 대푯값 추이 (선 그래프)', fontsize=14)
plt.xlabel('연도', fontsize=12)
plt.ylabel('임금', fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()
```
